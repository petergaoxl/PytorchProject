```python
import torch
```

```python
my_torch = torch.arange(10)
my_torch
```

```python
#  Reshape and View
my_torch = my_torch.reshape(2,5)
my_torch
```

```python
#  Reshape if we dont't know the number of items using -1
torch2 = torch.arange(15)
torch2
torch2 = torch2.reshape(3,-1)
torch2 = torch2.reshape(-1,5)
torch2
```

**View**

```python
torch3 = torch.arange(10)
torch3
```

```python
torch4 = torch3.view(2,5)
torch4
```

```python
#  What's the difference between shape and view?
torch5 = torch.arange(10)
torch5
```

```python
torch6 = torch5.view(2,5)
torch6
```
